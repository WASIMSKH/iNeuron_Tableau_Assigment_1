# IPL Data 2008 - 2020

```python
# Import Numpy and Pandas Library
import numpy as np
import pandas as pd
```

Create dataframe for IPL Ball by Ball

```python
df_ball = pd.read_csv("IPL Ball-by-Ball 2008-2020.csv")
```

```python
df_ball.head()
```

```python
df_ball.describe()
```

```python
df_ball.groupby("over").sum()
```

Create dataframe for IPL batsman

```python
df_match = pd.read_csv("IPL Matches 2008-2020.csv")
```

```python
df_match.head()
```

```python
df_match.describe()
```

```python
df_match.groupby("result").sum()
```

Join the dataframe on id column and this will give us the final data to analyze on Tableau

```python
df_final = pd.merge(df_ball, df_match, on="id", how="inner")
```

```python
df_final.head()
```

```python
df_final.columns
```

Export final output and visulaize on Tableau

```python
df_output = df_final.to_csv("df_final.csv", index=True)
```
